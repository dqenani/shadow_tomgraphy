## OSCAR Zeromode Analysis

This notebook is auxilary to main.py which calculates the zeromode of the Orenstein Uhlenbeck Operator with and without the use of Classical Shadows as run on the supercomputer OSCAR. The purpose of this notebook is to graph and analyze the results. 

```python
import csv
import numpy as np
import time
import matplotlib.pyplot as plt
import pandas as pd
```

```python
AER_SHADOW_PATH = "/Users/danielqenani/Desktop/Shadow-QPE/Results/Aer_Shadow_Results.csv" 
AER_NONSHADOW_PATH = "/Users/danielqenani/Desktop/Shadow-QPE/Results/Aer_NonShadow_Results.csv"
FM_SHADOW_PATH = "/Users/danielqenani/Desktop/Shadow-QPE/Results/FakeMontreal_Shadow_Results.csv"
FM_NONSHADOW_PATH = "/Users/danielqenani/Desktop/Shadow-QPE/Results/FakeMontreal_NonShadow_Results.csv"
np.set_printoptions(precision=6, suppress=True, linewidth=150)
```

```python
df_shadow = pd.read_csv(AER_SHADOW_PATH)
df_nonshadow = pd.read_csv(AER_NONSHADOW_PATH)

aer_snapshot_range = [int(i) for i in df_shadow.columns]

aer_shadow_data = df_shadow.to_numpy()[1:5,:]
aer_nonshadow_data = df_nonshadow.to_numpy()[1:5,:]
print(f"Shadow Data:\n {aer_shadow_data}\n")
print(f"Non-Shadow Data:\n {aer_nonshadow_data}")
```

```python
plt.errorbar(aer_snapshot_range, np.mean(aer_shadow_data, axis = 0), yerr = np.std(aer_shadow_data, axis = 0), label = 'With Shadows')
plt.errorbar(aer_snapshot_range, np.mean(aer_nonshadow_data, axis = 0), yerr = np.std(aer_nonshadow_data, axis = 0), label ='Without Shadows')
plt.title("Error Rate of Zero Mode Calculation with and without Classical Shadows (Aer Simulator)")
plt.xlabel("Number of snapshots")
plt.ylabel("Distance between Quantum and Classical")
plt.legend()
```
